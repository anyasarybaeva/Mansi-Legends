```python
from google.colab import files

# Загрузите файл с вашего компьютера
uploaded = files.upload()
```

```python
# Импортируем библиотеки для работы с регулярными выражениями
import re

# Укажите имя файла, который загрузили
filename = list(uploaded.keys())[0]

# Открываем и читаем содержимое файла
with open(filename, 'r', encoding='utf-8') as file:
    text = file.read()

cotext = text
print(text)  # Выводим оригинальный текст для проверки
```

```python
def replaceUTF16(text):
    result_string = re.sub('', 'а̄', text)
    result_string = re.sub('', 'ē', result_string)
    result_string = re.sub('', 'ё̄', result_string)
    result_string = re.sub('', 'о̄', result_string)
    result_string = re.sub('', 'ы̄', result_string)
    result_string = re.sub('', 'ю̄', result_string)
    result_string = re.sub('', 'э̄', result_string)
    result_string = re.sub('', 'я̄', result_string)

    result_string = re.sub('', 'А̄', result_string)
    # smallcase only
    # result_string = re.sub('\uf510', 'Ē', result_string)
    result_string = re.sub('', 'Ё̄', result_string)
    result_string = re.sub('', 'О̄', result_string)
    # smallcase only
    # result_string = re.sub('\uf520', 'Ы̄', result_string)
    result_string = re.sub('', 'Ю̄', result_string)
    result_string = re.sub('', 'Э̄', result_string)
    result_string = re.sub('', 'Я̄', result_string)

    # чистим переносы строк
    result_string = re.sub('(?<=[а̄ēё̄оы̄ю̄э̄я̄А̄ĒЁ̄О̄Ы̄Ю̄Э̄Я̄̄ӮӯӇӈ\w])[-–]\s(?![А-ЯА̄ĒЁ̄О̄Ы̄Ю̄Э̄Я̄̄ӮӇ])', '', result_string)
    result_string = re.sub('(?<=[а̄ēё̄оы̄ю̄э̄я̄А̄ĒЁ̄О̄Ы̄Ю̄Э̄Я̄̄ӮӯӇӈ\w])[-–]\s', '-', result_string)

    return result_string
```

```python
replaceUTF16(cotext)
```

```python
# Сохраняем результат в новый файл
new_filename = 'edited_file.txt'

with open(new_filename, 'w', encoding='utf-8') as file:
    file.write(replaceUTF16(cotext))

# Предлагаем скачать файл на компьютер
files.download(new_filename)
```

```python
for i in range (10):
  uploaded = files.upload()

  filename = list(uploaded.keys())[0]

  # Открываем и читаем содержимое файла
  with open(filename, 'r', encoding='utf-8') as file:
      text = file.read()

  cotext = text

  # Сохраняем результат в новый файл
  new_filename = str (i) + 'edited_file.txt'

  with open(new_filename, 'w', encoding='utf-8') as file:
      file.write(replaceUTF16(cotext))

  # Предлагаем скачать файл на компьютер
  files.download(new_filename)

  i = i + 1
```
